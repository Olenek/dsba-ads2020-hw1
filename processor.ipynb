```python
data_small = pd.read_csv("results/intersection.csv")
data_small = data_small.drop(data_small.columns[-1], axis=1)
```

```python
data_small.tail()
# Big numbers => convert to seconds
```

```python
data_small["karatsuba (ms)"] = data_small["karatsuba (ms)"].apply(lambda x: x/1000)
data_small["school (ms)"] = data_small["school (ms)"].apply(lambda x: x/1000)
```

```python
data_small = data_small.rename(columns={"karatsuba (ms)": "karatsuba (s)", "school (ms)": "school (s)"})
```

```python
fig_2 = data_small.plot(x="Problem size", figsize=(13, 13)).get_figure()
```

```python
fig_2.savefig('graphs/intersection.png', dpi=300)
```

```python
data_small["n^2 / 3 * 10^(-6)"] = data_small["Problem size"].apply(lambda x: x * x / 3000 / 1000)
data_small["n^2 / 3 * 10^(-7)"] = data_small["Problem size"].apply(lambda x: x * x / 3000 / 10000)
data_small["n^log_2(3) / 3 * 10^(-4)"] = data_small["Problem size"].apply(lambda x: (x**1.587) / 30 / 1000)
data_small["n^log_2(3) / 3 * 10^(-5)"] = data_small["Problem size"].apply(lambda x: (x**1.587) / 30 / 10000)
```

```python
fig_3 = data_small.plot(x="Problem size", y= ["n^2 / 3 * 10^(-6)", "n^2 / 3 * 10^(-7)", "school (s)"] , figsize=(10,5)).get_figure()
```

```python
fig_4 = data_small.plot(x="Problem size", y= ["n^log_2(3) / 3 * 10^(-4)", "n^log_2(3) / 3 * 10^(-5)", "karatsuba (s)"] , figsize=(10,5)).get_figure()
```

```python
data_all["n^2 / 3 * 10^(-1)"] = data_all["Problem size"].apply(lambda x: (x**2) / 30)
data_all["n^2 / 3 * 10^(-2)"] = data_all["Problem size"].apply(lambda x: (x**2) / 300)
```

```python
fig_5 = data_all.plot(x="Problem size", y= ["n^2 / 30", "n^2 / 3 * 10^(-2)", "caesar (ms)"] , figsize=(10, 5)).get_figure()
```

```python
fig_3.savefig('graphs/O_school.png', dpi=300)
fig_4.savefig('graphs/O_karatsuba.png', dpi=300)
fig_5.savefig('graphs/O_caesar.png', dpi=300)
```
